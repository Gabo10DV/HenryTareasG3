1) Abrir el archivo "Emisiones_CO2.csv" y cargar sus datos en un diccionario.

```python
import os
archivo = open('Emisiones_CO2.csv', 'r')
```

```python
dicc_emisiones = {  'cod_pais' : [],
                    'nom_pais' : [],
                    'region' : [],
                    'anio' : [],
                    'co2' : [],
                    'co2_percapita' : []}
```

```python


dicc_emisiones = {  'cod_pais' : [],
                    'nom_pais' : [],
                    'region' : [],
                    'anio' : [],
                    'co2' : [],
                    'co2_percapita' : []}
    
with open ('Emisiones_CO2.csv','r') as archivo_csv:
    next(archivo_csv) #saltar el encabezado
    for linea in archivo_csv:
              # Eliminar el salto de línea y dividir la línea por comas
        fila = linea.rstrip('\n').split("|")
        # Verificar que la fila tenga al menos 6 elementos
        if len(fila) >= 6:
            # Añadir los datos a la estructura
            dicc_emisiones['cod_pais'].append(fila[0])
            dicc_emisiones['nom_pais'].append(fila[1])
            dicc_emisiones['region'].append(fila[2])
            dicc_emisiones['anio'].append(int(fila[3]))

            #Reemplazar coma por punto 
            co2_co2 = fila[4].replace(',','.')
            dicc_emisiones['co2'].append(float(co2_co2))

            # Reemplazar coma por punto en el último dato de la columna 'co2_percapita'
            co2_percapita = fila[5].replace(',', '.')
            dicc_emisiones['co2_percapita'].append(float(co2_percapita))
        else:
            print("La fila no tiene suficientes elementos:", fila)
```

```python
""" para saber cuantas variables hay se cuentas las columnas
es decir la cantidad de keys que tiene el diccionario"""
print(len(dicc_emisiones.keys()))
```

```python
"""para calcular el total de emisiones primero hay que convertir el tipo de veriables
que el ser en csv se toman como texto"""

emisiones_totales = 0
for i in range(len(dicc_emisiones['co2'])):
    if dicc_emisiones['region'] == 'America Latina y Caribe' and dicc_emisiones['anio']==2010:
        emisiones_totales += dicc_emisiones['co2'][i]

    print(emisiones_totales)
```
